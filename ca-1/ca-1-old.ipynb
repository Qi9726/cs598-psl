## Part 1: Generate Data

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier as knn
from sklearn import linear_model as lm

np.random.seed(100)

```

First generate the 20 centers from two-dimensional normal . You can use any mean and covariance structure.  You should not regenerate the centers. Use these 20 centers throughout this simulation study.

```python
# Generate 20 centers - TBD

center_count = 20

p = 2
s = 1      # sd for generating x       
m1 = [1,0]
m0 = [0,1]
m2 = [2,0]
m3 = [0,2]
centers = np.array([m1, m0, m2, m3])
centers = np.repeat(centers, repeats=5, axis=0)
centers.shape
```

Given the 20 centers, generate a training sample of size 200 (100 from each class) and a test sample of size 10,000 (5,000 from each class).

```python
n = 100  # generate the train data; n for each class

repeat_centers = np.repeat(centers, repeats=(2*n/center_count), axis=0)
traindata = np.random.normal(size = (2*n, p)) * s + repeat_centers
Ytrain = np.concatenate(([1]*n, [0]*n))
traindata.shape
```

```python
N = 5000  # generate the test data; N for each class


repeat_centers = np.repeat(centers, repeats=(2*N/center_count), axis=0)
testdata = np.random.normal(size = (2*N, p)) * s + repeat_centers
Ytest = np.concatenate(([1]*N, [0]*N))
testdata.shape
```

Visualization

```python
plt.scatter(traindata[:n, 0], traindata[:n, 1], c = "blue", alpha=0.2, label='Class 1')
plt.scatter(traindata[n:, 0], traindata[n:, 1], c = "red", alpha=0.2, label='Class 0')

#plt.scatter(m1[0], m1[1], marker = '+', s = 80, c = "blue")
#plt.scatter(m0[0], m0[1], marker = '+', s = 80, c = "red")

c_n = int(center_count/2)
plt.scatter(centers[:c_n, 0], centers[:c_n, 1], marker = '+', s = 80, c = "blue")
plt.scatter(centers[c_n:, 0], centers[c_n:, 1], marker = '+', s = 80, c = "red")

plt.legend()
plt.show()
```
